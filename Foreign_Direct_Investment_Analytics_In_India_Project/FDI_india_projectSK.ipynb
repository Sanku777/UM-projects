## Visualization :

## Exploratory Data Analysis(EDA)

# Year-wise total Investment

```python
FDI_Data.set_index('Sector', inplace=True)
```

```python
#Creation of Barplot for visualization

FDI_Data.sum(axis=0).plot(kind='bar', color='g', figsize=(10,5))
plt.title('Total FDI in India (2000-01 to 2016-17)')
plt.xlabel('Financial Year')
plt.ylabel('Total FDI (in million USD)')
plt.xticks(rotation = 45)
plt.show()
```

## So using bar plot, we can say that the highest total investment was done in the year of 2016-17. And in the year of 2000-01,2003-04 was lowest investment.



```python
#Creation of lineplot using matplot and seaborn library.

plt.figure(figsize=(10,5))
sns.lineplot(data = FDI_Data.sum(axis=0))
plt.xticks(rotation = 45)
plt.title('Total FDI in India (2000-01 to 2016-17)')
plt.xlabel('Financial Year')
plt.ylabel('Total FDI (in million USD)')
plt.show()
```

## So, We can see in this plot that total investment is continuously increasing in every year.

# Sector-wise total Investment

```python
FDI_Data.sum(axis=1).sort_values().plot(kind = 'bar', figsize=(15,5))
plt.title('Sector-wise total investment')
plt.xlabel('Sector')
plt.ylabel('Total FDI (in million USD)')
plt.xticks(rotation = 90)
plt.show()
```

## Here we can see that the total FDI of the sectors. Services sector is the highest sector, who was invest highest total of FDI in India. Their is very low investment towards sectors like Coir, Defence Industries, Mathematical,surveying and drawing Instruments, Coal Production. There are around 10 to 15 sectors, investment is very less and some of this like zero.

# Now adding a new column into our dataset for more exploration :

```python
FDI_Data['Category Total'] = FDI_Data.sum(axis = 1)                              # Adding new column 'Category Total' for working perpose.
FDI_Data.sort_values(by = 'Category Total', ascending=False, inplace = True)
FDI_Data.head()
```

# Looking about top 5 categorries :

```python
top5_categories = FDI_Data.head()
top5_categories = pd.DataFrame(top5_categories['Category Total'])
```

```python
top5_categories
```

# Visualise percentage share of FDI among top 5 sectors :

```python
#Creating pie chart to visualise percentage share of FDI among top 5 sectors using Matplotlib

plt.figure(figsize=(20,8))
plt.pie(top5_categories['Category Total'],labels=top5_categories.index,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('SHARE AMONG TOP 5 SECTORS WITH HIGHEST FDI')
plt.show()
```

## Here we can see in the above Chart that Services Sector Managed to Attract highest percentage share of FDI greater than any other Sector and among top 5 Sectors it has 39.9% share. Computer software and hardware has 16.6%. Construction development has occupied 16.3%, Telecommunications 16.1 and Automobile industry 11.2%.

# Data transformation 

```python
top5_categories = top5_categories.transpose()
```

```python
top5_categories.rename(columns={'SERVICES SECTOR (Fin.,Banking,Insurance,Non Fin/Business,Outsourcing,R&D,Courier,Tech. Testing and Analysis, Other)': 'SERVICES SECTOR', 'CONSTRUCTION DEVELOPMENT: Townships, housing, built-up infrastructure and construction-development projects': 'CONSTRUCTION DEVELOPMENT'}, inplace = True)
```

# We have rename some column name for better visualization.

# Visualization of Top 5 companies with highest FDI :

## Using Barplot

```python
plt.figure(figsize=(8,6))
sns.barplot(data = top5_categories)
plt.xticks(rotation = 80)
plt.title('Top 5 Companies with highest FDI')
plt.ylabel('Total FDI')
plt.show()
```

# Here we can see that the top foreign direct investment in India from service sector then computer software and hardware and others.
